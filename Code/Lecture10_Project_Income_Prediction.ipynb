```python
import pandas as pd
import numpy as np
```

```python
url = "https://archive.ics.uci.edu/ml/machine-learning-databases/adult/adult.data"
Income_Data = pd.read_csv(url, header = None, names = 
                          ['age', 'workclass', 'nlwgtf', 'education', 'education_num', 
                           'marital_status', 'occupation', 'relationship', 'race', 'sex', 
                           'capital_gain', 'capital_loss', 'hours_per_week', 'native_country', 'MoreThan50']

                         )
```

You can find more info about this dataset [here](https://archive.ics.uci.edu/ml/datasets/Census+Income)

```python
Income_Data.head()
```

```python
Income_Data.describe()
```

```python
Income_Data.dropna(how = 'any', inplace = True)
Income_Data.describe()
```

```python
workclass_dummy = pd.get_dummies(Income_Data.workclass, prefix = 'work')
del workclass_dummy[workclass_dummy.columns.values[0]]

sex_dummy = pd.get_dummies(Income_Data.sex, prefix = 'Sex')
del sex_dummy[sex_dummy.columns.values[0]]

mar_status_dummy = pd.get_dummies(Income_Data.marital_status, prefix = 'mar_status')
del mar_status_dummy[mar_status_dummy.columns.values[0]]

occupation_dummy = pd.get_dummies(Income_Data.occupation, prefix = 'occupation')
del occupation_dummy[occupation_dummy.columns.values[0]]

relationship_dummy = pd.get_dummies(Income_Data.relationship, prefix = 'relationship')
del relationship_dummy[relationship_dummy.columns.values[0]]

race_dummy = pd.get_dummies(Income_Data.race, prefix = 'race')
del race_dummy[race_dummy.columns.values[0]]

country_dummy = pd.get_dummies(Income_Data.native_country, prefix = 'country')
del country_dummy[country_dummy.columns.values[0]]

morethan50_dummy = pd.get_dummies(Income_Data.MoreThan50, prefix = 'over_50')
del morethan50_dummy[morethan50_dummy.columns.values[0]]

education_dummy = pd.get_dummies(Income_Data.education, prefix = 'education')
del education_dummy[education_dummy.columns.values[0]]


Income_Data = pd.concat([Income_Data, workclass_dummy, sex_dummy, mar_status_dummy, occupation_dummy, relationship_dummy, race_dummy,
                        country_dummy, morethan50_dummy, education_dummy], axis=1)

Income_Data.head()
```
