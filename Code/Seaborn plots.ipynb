```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline

iris = sns.load_dataset("iris")
sns.pairplot(iris);
```

```python
sns.jointplot(x='sepal_width', y='sepal_length', data=iris);
```

```python
url = "https://archive.ics.uci.edu/ml/machine-learning-databases/00242/ENB2012_data.xlsx"
DE = pd.read_excel(url, names = ['Relative_Compactness', 'Surface_Area', 'Wall_Area','Roof_Area', 
                                          'Overall_Height', 'Orientation', 'Glazing_Area', 'Glazing_Area_Distribution',
                                          'y_Heating_Load', 'y_Cooling_Load'])
sns.pairplot(DE);
```

```python
sns.jointplot(y='y_Cooling_Load', x='Orientation', data=DE);
```

```python
sns.set_style("white")
```

```python
x = DE.y_Cooling_Load
fig, (ax1, ax2) = plt.subplots(ncols=2, nrows=1)
sns.distplot(x, ax=ax1, color='red', )
sns.distplot(x, kde=False, ax=ax2)
fig.set_size_inches(12,6)
ax1.set_title('Red Graph', color='red')
ax1.set_axis_bgcolor(color='aliceblue')
ax1.grid(color='gray',zorder=-1)


plt.show()
```

```python
sns.set_style("dark",{"axes.facecolor": "antiquewhite"})

# multiple ways to generate the graph -- version 1
DE.y_Cooling_Load.hist()
plt.show()
# more styling possibilities at https://stanford.edu/~mwaskom/software/seaborn/tutorial/aesthetics.html
```

```python
sns.set_style("dark") #this style hides the grid because the grid matches the facecolor

plt.hist(DE.y_Cooling_Load) # version 2
plt.show()
```

```python

fig, (ax3, ax4) = plt.subplots(ncols=2, nrows=1)
fig.set_size_inches(15,8)
gender = ['male','male','female','whatever','female', 'male']
#plt.sca(ax3)
sns.barplot([1,2,3,4,5,6], [10,14,9,11,14,10], ax=ax3)
#plt.sca(ax4)
sns.barplot([10,14,9,11,14,10],[1,2,3,4,5,6],  hue=gender, ax=ax4, orient='h', palette=('cadetblue','orange','olive'))
sns.despine()
plt.show()
```

```python
#countplot 
a = np.random.randint(0,2, size=100)
b = np.random.randint(0,5, size=100)
cats = pd.DataFrame(data = {'a':a,'b':b})
cats.describe()

ages = ['baby', 'child', 'teen', 'adult', 'senior']

cats['gender'] = ['male' if x==0 else 'female' for x in cats['a']]
cats['age'] = [ages[x] for x in cats.b]

print cats.head()
plt.figure(figsize=(12, 7))
plt.ylim(0, 18)

sns.set(style="darkgrid")
ax = sns.countplot(x="gender", hue="age", data=cats)

```

```python

plt.xlim(0, 18)

ax = sns.countplot(y="gender", hue="age", data=cats, orient='h')
plt.show()
```

```python
data = pd.DataFrame(np.random.multivariate_normal([0,0], [[1,-0.5],[-0.5,1]], size=100), columns = ['one','two'], )

print data.describe()
# plot
sns.set_style('ticks')
fig, ax = plt.subplots()
fig.suptitle('demo linear regression', fontsize=30)
fig.set_size_inches(18.5, 10.5)
sns.regplot(data.one, data.two, ax=ax)
sns.despine()  #get rid of axes along the top

ax.set_xlabel('Across', fontsize=16) #must come after graph is calculated
ax.set_ylabel('---- Up --->', fontsize=16, fontweight='bold') #must come after graph is calculated

plt.show()
```
